    4. Confeccionar los siguientes Gráficos

        Histograma: Un histograma es útil para ver la distribución de una variable, es decir, nos permite ver los valores más comunes. Para el caso que se presenta se desea ver la distribución de la cantidad media de habitaciones en el estudio realizado.

```python
#Importar librerias
import numpy as np
fig,ax = plt.subplots()
#cantidad media de habitaciones
x = df["RM"]
ax.hist(x, color = "tab:blue")
#Titulos
plt.title('Grafico de cantidad media de habitaciones')
plt.xlabel('x: Cantidad media de habitaciones')
plt.ylabel('y: Frecuencia de habitaciones')
plt.show()
```

        Dispersión: sirve para representar la relación entre dos variables. Para nuestro caso se desea ver la relación entre índice de crimen y el valor mediano de las casas. En otras palabras, ¿el índice de crimen afecta el valor medio de las casas?

```python
#Importar librerias
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
#relación entre índice de crimen y el valor mediano de las casas
ax.scatter(df["CRIM"], df["MEDV"],  color = "tab:blue")
#Titulos
plt.title('Grafico relación entre índice de crimen y el valor mediano de las casas')
plt.xlabel('x: índice de crimen')
plt.ylabel('y: valor mediano de las casas')
plt.show()
```

        Barras: Un gráfico de barras es útil para comparar una variable entre distintos grupos o categorías. Por ejemplo, si quisiéramos observar el valor medio de cada ciudad. Para nuestro caso se desea agrupar por ciudad y toma la media del valor mediano. Gráfica las 10 primeras ciudades.

```python
#Unir los valores de la cuidad y valores medios
df1 = df.groupby(['TOWN']).agg({'MEDV': 'mean'})
df2 = pd.DataFrame()
df2['TOWN'] = df1.index
df2['MEDV'] = df1.values
#Hacer una lista solo con los 10 primeros valores
df3 = df2.head(10)
#Grafico de barras
fig, ax = plt.subplots()
ax.barh( df3['TOWN'], df3['MEDV'] , color = "tab:blue")
plt.title('Grafico que agrupa por ciudad y toma la media del valor mediano')
plt.xlabel('x: valor mediano de las casas')
plt.ylabel('y: 10 primeras cuiudades')
plt.show()
```
