a) Generar un DataFrame con los datos del fichero.

```python
#Importar
import pandas as pd
#Dividir en punto y coma ademas de considerar el punto como decimal. Y mostar. 
df = pd.read_csv("titanic.csv", sep=';', decimal='.')
display(df)
```

e) Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas alfabéticamente.

```python
#Filtrar la primera clase y ordenar alfabeticamente
print(df[df["Pclass"]==1]['Name'].sort_values())
```

f) Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.

```python
#Mostrar porcentaje que sobrevivio y murio
a = df['Survived'].value_counts()/df['Survived'].count() * 100
print(a)
#Mostrar
print(f"\n")
print("Sobrevivieron 38.38% y murireron 61.61%")
```

g) Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.

```python
#Mostrar porcentaje que sobrevivio por clase
a1 = df.groupby('Pclass')['Survived'].value_counts(normalize=True)
print(a1)
#Mostrar
print(f"\n")
print("En la clase 1: Sobrevivieron 0.62% y murireron 0.37%")
print("En la clase 2: Sobrevivieron 0.47% y murireron 0.52%")
print("En la clase 3: Sobrevivieron 0.24% y murireron 0.75%")
```

h) Eliminar del DataFrame los pasajeros con edad desconocida.

```python
#Primera froma de eliminar edades desconocidas
newdf = df['Age'].dropna()
newdf
```

```python
#Segunda froma de eliminar edades desconocidas
df = df[df['Age'].notna()]
df
```

i) Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.

```python
#Mostrar mujeres que viajaban por clase
print(df.groupby(['Pclass','Sex'])['Age'].mean().unstack()['female'])
#Mostrar
print(f"\n")
print("En la clase 1: La edad promedio de las muejres era de 34")
print("En la clase 2: La edad promedio de las muejres era de 28")
print("En la clase 3: La edad promedio de las muejres era de 21")
```

j) Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.

```python
#Nueva columna boleana
df['J'] = df['Age'] < 18
#Mostar
print(df['J'])
```

k) Mostrar por pantalla el porcentaje de menores y mayores de edad que sobrevivieron en cada clase.

```python
#Mostrar por clase mayores y menores que sobrevivieron
print(df.groupby(['Pclass', 'J'])['Survived'].value_counts(normalize = True) * 100)
#Mostrar
print(f"\n")
print("En la clase 1: El porcentaje de menores de edad que sobrevivieron son 91.66%")
print("En la clase 1: El porcentaje de menores de edad que no sobrevivieron son 8.33%")
print("En la clase 1: El porcentaje de mayores de edad que sobrevivieron son 63.79%")
print("En la clase 1: El porcentaje de mayores de edad que no sobrevivieron son 36.20%")
print(f"\n")
print("En la clase 2: El porcentaje de menores de edad que sobrevivieron son 41.33%")
print("En la clase 2: El porcentaje de menores de edad que no sobrevivieron son 58.66%")
print("En la clase 2: El porcentaje de mayores de edad que sobrevivieron son 91.30%")
print("En la clase 2: El porcentaje de mayores de edad que no sobrevivieron son 8.69%")
print(f"\n")
print("En la clase 3: El porcentaje de menores de edad que sobrevivieron son 20.21%")
print("En la clase 3: El porcentaje de menores de edad que no sobrevivieron son 79.78%")
print("En la clase 3: El porcentaje de mayores de edad que sobrevivieron son 37.17%")
print("En la clase 3: El porcentaje de mayores de edad que no sobrevivieron son 62.82%")
```
