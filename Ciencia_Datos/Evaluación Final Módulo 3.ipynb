Crear un programa utilizando la librería que realice lo siguiente:
1. Crear el siguiente DataFrame indexado:

```python
#Librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Para hcer la tabla
valores = { "Calorias": [420, 380, 390, 490, 300],
          "Tiempo": [60, 40, 75, 55, 45]}
#Index
indice = ["L", "M", "X", "J", "V"]

df = pd.DataFrame(valores, indice)
print(df)
```

2. Calcular la media, mediana, y desviación típica de ambas columnas.

```python
#Media
print('El valor de la media es ' , round(df.mean(), 0))
```

```python
# Mediana
print('El valor de la mediana es ', round(df.median(), 0))
```

```python
#Desviación tipica
print('El valor de la desviacion tipica', round(df.std(), 0))
```

3. Añadir otra columna booleana al DataFrame para ver si se ha cumplido el reto de quemar más de 400 calorías por hora. La nueva
columna debe generarse aplicando una fórmula a las otras columnas.
El DataFrame resultante debe ser el siguiente:

```python
#Hacer el verdadero y falso
a = (400 / 60)
b = (df.Calorias / df.Tiempo)
df["Booleano"] = b > a
#Mostar tabla
print(df)
```

4. Filtrar el DataFrame y devolver otro DataFrame con las filas pares que cumplan que el número de calorías es mayor de 400.

```python
#Filtrar por 400
a1 = df.iloc[range(0, df.shape[0], 2)]
a1 = a1[a1.Calorias > 400]
print(a1)
```

5. Crear a partir del DataFrame una serie con los porcentajes de días que se ha conseguido el reto y los que no.

```python
#Porcentaje de booleano
c = df.Booleano.value_counts(normalize = True)
print(c * 100)
```

6. Crear un gráfico como el de más abajo que muestre la progresión de las calorías y tiempo durante la semana.

```python
#Graficar
fig, ax = plt.subplots()
df.plot(y = 'Calorias', ax = ax)
df.plot(y = 'Tiempo', ax = ax)
#Titulos
plt.title('Grafico progresión de las calorías y tiempo durante la semana')
plt.xlabel('x: tiempo durante la semana')
plt.ylabel('y: progresión de las calorías')
plt.show()
```
