Caso: A partir del fichero del Titanic, realizar los siguiente:

1. Confeccione un gráfico de barras para visualizar hombres y mujeres que han sobrevivido.


```python
#Importar
import pandas as pd
#Dividir en punto y coma ademas de considerar el punto como decimal. Y mostar. 
df = pd.read_csv("titanic.csv", sep=';', decimal='.')
display(df)
```

```python
#Filtro solicitado
df1 = df.groupby(['Sex','Survived'])['Survived'].sum().unstack()[1]
print(df1)
```

```python
#Importar
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#Grafico de barras
fig, ax = plt.subplots()
ax.barh( df1.index, df1)
#Titulos
plt.title('Grafico que  visualizar hombres y mujeres que han sobrevivido')
plt.xlabel('x: cantidad de personas que han sobrevivido')
plt.ylabel('y: sexo de las personas')
plt.show()
```

2. Cambiar el color del gráfico.

```python
#Importar
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#Grafico de barras
fig, ax = plt.subplots()
ax.barh( df1.index, df1 , color = "tab:red")
#Titulos
plt.title('Grafico que  visualizar hombres y mujeres que han sobrevivido')
plt.xlabel('x: cantidad de personas que han sobrevivido')
plt.ylabel('y: sexo de las personas')
plt.show()
```

3. Gráfico barras agrupados

```python
#Filtro solicitado solo con los sobrevivientes
df2 = df.groupby(['Sex','Pclass','Survived'])['Survived'].sum().unstack()[1]
print(df2)
```

```python
#Importar
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#Grafico de barras por las clases
df2.groupby(["Pclass", "Sex"]).sum().unstack().plot(kind = "bar")
#Titulos
plt.title('Grafico que  visualizar hombres y mujeres que han sobrevivido, por clase')
plt.xlabel('x: tipo de clase')
plt.ylabel('y: cantidad de personas que han sobrevivido')
plt.show()
```
